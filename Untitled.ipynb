```python
import tensorflow as tf
```

```python
x_data=[10.0,9.0,3.0,2.0]
y_data=[90.0,80.0,50.0,30.0]

x=tf.placeholder(tf.float32)
x2=x**2
x3=x**3

y=tf.placeholder(tf.float32)
w0=tf.Variable(tf.random_normal([1]))
w1=tf.Variable(tf.random_normal([1]))
w2=tf.Variable(tf.random_normal([1]))
w3=tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]), name ='bias')
hy=w0+w1*x+w2*x2+w3*x3 +b

cost=tf.reduce_mean(tf.square(y-hy))
opt=tf.train.GradientDescentOptimizer(learning_rate=1e-6)
train=opt.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for a in range(0,100):
        cost_val,hy_val,_=sess.run([cost,hy,train],feed_dict={x:x_data, y:y_data})
        if a%1==0:
            print(cost_val,hy_val)
    resu=sess.run([w0,w1,w2,w3])
    print(resu)
```
